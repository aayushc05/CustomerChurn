```python
#remove unnecessary variables from df
final_df=df[['TotalCharges','tenure','MonthlyCharges','Contract','InternetService','PaymentMethod','Churn']]
final_df=pd.get_dummies(final_df,drop_first=True)
final_df.columns = final_df.columns.str.replace(' ', '_') 
final_df.columns = final_df.columns.str.replace('(', '')
final_df.columns = final_df.columns.str.replace(')', '') 
final_df.head()
```

```python
X=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Randomized Search CV

import numpy as np

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt', 'log2']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestClassifier()
```

```python

# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='roc_auc', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
best_random=rf_random.best_estimator_
```

```python
rf_random.best_params_
```

```python
predictions=best_random.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```
